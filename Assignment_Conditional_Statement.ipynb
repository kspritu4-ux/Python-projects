```python
# Question 1 : Write a program that takes the current temperature and provides clothing advice

# Ask user for current temperature
temperature = float(input("Enter the current temperature in Celsius: "))

# Provide clothing advice based on temperature
if temperature < 0:
    advice = "It's freezing! Wear a heavy coat, gloves, and a warm hat."
elif 0 <= temperature < 10:
    advice = "Very cold. Bundle up with a coat and scarf."
elif 10 <= temperature < 20:
    advice = "Cool weather. A light jacket or sweater should be fine."
elif 20 <= temperature < 30:
    advice = "Warm. Dress comfortably—t-shirt and jeans work well."
elif 30 <= temperature < 40:
    advice = "Hot! Wear breathable clothes like shorts and a tank top."
else:
    advice = "Scorching! Stay hydrated, wear minimal clothing, and avoid direct sun."

# Output the advice
print(f"Temperature: {temperature}°C")
print("Clothing Advice:", advice)
```

```python
# Question 2 : Create a program that determines if an applicant qualifies for a loan based on their
# salary and years at their current job.

salary = float(input("Salary (₹): "))
years = float(input("Years at job: "))

if salary >= 30000 and years >= 2:
    print("Congratulations! Loan approved")
elif salary >= 30000 and years <= 2:
    print("Oops! Loan denied because year is not fulfiling the requirements")
elif salary <= 30000 and years >= 2:
    print("Sorry! Loan denied because your salary is unable to meets our requiements")
else:
    print("please! enter valid inputs")
```

```python
# Question 3 : Develop a script that assigns grades to students based on their score.

score = float(input("Enter the student's score: "))

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"Grade: {grade}")
```

```python
#Question 4 :  Write a Python script that asks for the number of items and the price per item,
# then calculates and prints the total cost with a 10% discount for 10 or more items, or a 20% discount for 20 or more items.

# Input
items = int(input("Enter number of items: "))
price = float(input("Enter price per item (₹): "))

# Calculate total cost
total = items * price

# Apply discount
if items >= 20:
    discount = 0.20
elif items >= 10:
    discount = 0.10
else:
    discount = 0.0

final_total = total * (1 - discount)

# Output
print(f"Total cost after discount: ₹{final_total:.2f}")
```

```python
# Question 5 : Write a Python script that inputs the time (24-hour format) and the customer's age, then prints the ticket price.
# Ticket prices are as follows:
# Regular hours (before 5 PM): $12 for adults, $8 for seniors (60+), $5 for children (under 12).
# Evening hours (5 PM and later): $15 for adults, $10 for seniors, $8 for children.


# Input
time = int(input("Enter time (0–23): "))
age = int(input("Enter age: "))

# Determine pricing
if time < 17:  # Regular hours
    if age < 12:
        price = 5
    elif age >= 60:
        price = 8
    else:
        price = 12
else:  # Evening hours
    if age < 12:
        price = 8
    elif age >= 60:
        price = 10
    else:
        price = 15

# Output
print(f"Ticket price: ${price}")
```
